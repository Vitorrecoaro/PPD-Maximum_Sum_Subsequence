## Introdução

Na disciplina de programação paralela e distribuída, foi proposto um problema para cada grupo resolver com implementações paralelas utilizando as seguintes tecnologias:

- Open MP.
- MPI.
- CUDA.

E o problema que será abordado aqui é o problema de subsequência crescente de soma máxima, em que é proposto três valores de entrada:

- **n**: A quantidade de elementos que será lida.
- **k**: A quantidade de elementos que a subsequência deverá ter.
- **arr**: Um vetor com *n* elementos que será extraída a subsequência crescente de soma máxima. 

Assim como um exemplo, caso temos:

- **n** = 8
- **k** = 3
- **arr** = [8, 5, 9, 10, 5, 6, 21, 8]

Temos que o valor que deverá ser retornado é **40**, pois os 3 elementos selecionados para fazer a soma são: 9 + 10 + 21 = 40.

## Programa sequencial

A resolução do problema sequencial é feita com programação dinâmica, utilizando uma matriz **dp** de tamanho $n\times k + 1$, assim na coluna $j$ ele irá guardar a maior soma parcial que o elemento consegue fazer com $j$ elementos incluindo a ele mesmo. 

Deste modo o algoritmo irá iterar por todo o vetor completando essa matriz **dp**, indo de linha em linha, completando as colunas com as somas parciais com os dados já calculados das linhas anteriores. Assim após a última iteração é verificado na última coluna da matriz a maior soma parcial com **k** elementos, se existir, e assim retornando o valor para a função principal mostrando ao usuário este valor, caso o valor não exista o valor retornado é 0.

```python
%%writefile sequencial.c
#include <stdio.h>
#include <stdlib.h>
#define MAX(a, b) a > b ? a : b

int MaxIncreasingSub(int *arr, int n, int k)
{
	int **dp, ans = -1;
	dp = malloc(n * sizeof(int *));
	for (int i = 0; i < n; i++)
		dp[i] = malloc((k + 1) * sizeof(int));
	for (int i = 0; i < n; i++)
	{
		for (int j = 0; j < k; j++)
		{
			dp[i][j] = -1;
		}
	}
	for (int i = 0; i < n; i++)
	{
		dp[i][1] = arr[i];
	}

	for (int i = 1; i < n; i++)
	{
		for (int j = 0; j < i; j++)
		{
			if (arr[j] < arr[i])
			{
				for (int l = 1; l <= k - 1; l++)
				{
					if (dp[j][l] != -1)
					{
						dp[i][l + 1] = MAX(dp[i][l + 1], dp[j][l] + arr[i]);
					}
				}
			}
		}
	}

	for (int i = 0; i < n; i++)
	{
		if (ans < dp[i][k])
			ans = dp[i][k];
	}

	return (ans == -1) ? 0 : ans;
}

int main()
{
	int n, k;
	scanf(" %d", &n);
	scanf(" %d", &k);
	int *arr = malloc(n * sizeof(int));
	for (int i = 0; i < n; i++)
		scanf(" %d", &arr[i]);
	int ans = MaxIncreasingSub(arr, n, k);
	printf("%d\n", ans);
	return 0;
}
```

## Programa paralelo com Open MP

Antes de pensar em paralelizar, é possível observar a grande dependência de dados no código sequencial, sendo necessário criar uma versão linearizada, a qual será utilizada em todas paralelizações deste trabalho.

Após a mudança, implementamos a paralelização do OpenMP, através, inicialmente, de parallel for na inicialização do dp[i] como -1 em toda matriz e no For seguinte a atribuição de elementos do array na primeira linha da matriz dp.

Na parte principal do código, foi preciso utilizar um private em "l" para impedir que os dados dependessem de outras threads, caso contrário uma thread teria que esperar o término dos cálculos anteriores para prosseguir.

Por último, para fazer a soma final do valor obtido, adicionamos a diretiva For reduction para que os resultados de cada threads sejam independentes, e, após os cálculos, todas somas parciais irão ser somadas para chegar ao resultado final.

```python
%%writefile omp-prog.c
#include <stdio.h>
#include <stdlib.h>
#define MAX(a, b) a > b ? a : b

int MaxIncreasingSub(int *arr, int n, int k)
{
	int *dp = (int *) malloc(n * k * sizeof(int));
	int ans = -1;

	int i, l, j;

#pragma omp parallel for
	for (i = 0; i < n * k; i++)
	{
		dp[i] = -1;
	}

#pragma omp parallel for
	for (i = 0; i < n; i++)
	{
		dp[i] = arr[i];
	}

	for (int i = 0; i < (k - 1); i++)
	{
#pragma omp parallel for private(l)
		for (int j = i + 1; j < n; j++)
		{
			for (int l = 0; l < j; l++)
			{
				if (arr[l] < arr[j])
				{
					if (dp[i * n + l] != -1)
					{
						dp[(i + 1) * n + j] = MAX(dp[(i + 1) * n + j], dp[i * n + l] + arr[j]);
					}
				}
			}
		}
	}

#pragma omp parallel for reduction(max: ans)
	for (i = 0; i < n; i++)
	{
		if (ans < dp[(k - 1) * n + i])
			ans = dp[(k - 1) * n + i];
	}

	free(dp);
	return (ans == -1) ? 0 : ans;
}

int main()
{
	int n, k;
	scanf(" %d", &n);
	scanf(" %d", &k);
	int *arr = malloc(n * sizeof(int));
	for (int i = 0; i < n; i++)
		scanf(" %d", &arr[i]);
	int ans = MaxIncreasingSub(arr, n, k);
	printf("%d\n", ans);
	return 0;
}
```

## Programa paralelo com CUDA

Na paralelização com CUDA, pelo motivo da GPU possuir grandes quantidades de unidades de processamento, iremos atribuir a cada unidade de processamento apenas uma soma parcial. Assim, sendo ativado `k - 1` vezes a função que calcula a soma parcial e armazena na linha de interesse, para que na próxima rodada este valor possa ser utilizado.

`calculatePartialSum`: Esta função calcula a soma parcial necessária para obter a subsequência crescente mais longa de comprimento `k`. A função recebe cinco argumentos: 
* `totalOperations` é a quantidade total de operações que serão realizadas; 

* `lin` é a linha atual do array `dp` que está sendo calculada; 

* `n` é o tamanho do vetor de entrada; 

* `arr` é o vetor de entrada; 

* `dp` é o vetor de saída.

A função primeiro calcula o índice da thread que está executando, em seguida, calcula o índice correspondente no vetor de entrada `arr` e no vetor de saída `dp`. A função usa uma fórmula para calcular qual elemento do vetor a thread atual deve comparar com o elemento de referência. Se o elemento atual for menor que o elemento de referência, o valor do vetor `dp` é atualizado. A função usa **atomicMax** para garantir que várias threads não atualizem a mesma posição do vetor `dp` ao mesmo tempo.

O argumento totalOperations indica a quantidade total de operações que serão realizadas no kernel. Isso é calculado com base na posição atual (lin), no tamanho do vetor de entrada (n) e na quantidade de somas parciais a serem calculadas. Esse cálculo é feito usando uma fórmula que representa a soma dos elementos de uma Progressão Aritmética de ordem 1, que começa em lin e vai até n-1.

`calculateAnswer`: Esta função calcula a resposta final para o problema k-LIS. A função recebe quatro argumentos: 

* `n` é o tamanho do vetor de entrada; 

* `k` é o comprimento da subsequência crescente mais longa que estamos procurando; 

* `ans` é um ponteiro para o valor da resposta final; 

* `dp` é o vetor de saída.

A função calcula o índice da thread que está executando e, em seguida, verifica se o valor atual do vetor `dp` é maior do que o valor atual de `ans`. Se for o caso, o valor de ans é atualizado. A função usa **atomicMax** para garantir que várias threads não atualizem a mesma posição do vetor `ans` ao mesmo tempo.

`MaxIncreasingSub`: Esta é a função principal que chama as funções `calculatePartialSum` e `calculateAnswer` para calcular a resposta final para o problema k-LIS. A função recebe três argumentos: 
* `arr` é o vetor de entrada; 
* `n` é o tamanho do vetor de entrada;
* `k` é o comprimento da subsequência crescente mais longa que estamos procurando. 

A função aloca memória na GPU para os vetores `d_arr` e `d_dp`, copia o vetor de entrada `arr` para a GPU e inicializa o vetor de saída `dp` com -1. Em seguida, a função chama a função `calculatePartialSum` para calcular as somas parciais necessárias. Depois que todas as somas parciais são calculadas, a função chama a função `calculateAnswer` para calcular a resposta final. A função retorna o valor de `ans`, que é a resposta final.

`main`: Esta é a função principal que lê a entrada do usuário, chama a função `MaxIncreasingSub` e imprime o resultado na tela. A função lê o tamanho do vetor de entrada `n` e o comprimento da subsequência crescente mais longa `k`. Em seguida, a função lê os elementos do vetor de entrada e chama a função `MaxIncreasingSub` para calcular a resposta final. Finalmente, a função imprime o resultado na tela.

```python
%%writefile cuda-prog.cu
#include<cuda.h>
#include<stdio.h>
#include<stdlib.h> 

__global__ void calculatePartialSum( int totalOperations, int lin, int n, int *arr, int *dp){
    int idx = threadIdx.x + blockIdx.x * blockDim.x;

    if(idx < totalOperations){
        int refValue = (int)(1 + sqrtf((float)(1 - 4 * ((lin + 1) - ((lin + 1) * (lin + 1)) - 2 * idx)))) / 2;
        int compValue = (int)(idx) - (int)(refValue * refValue + (lin + 1) - refValue - (lin + 1) * (lin + 1)) / 2;

        if(arr[compValue] < arr[refValue]){
            if(dp[lin * n + compValue] != -1){
                int newVal = dp[lin * n + compValue] + arr[refValue];
                atomicMax(&dp[(lin + 1) * n + refValue ], newVal);
            }
        }
    }
}

__global__ void calculateAnswer(int n, int k, int *ans, int *dp){
    int idx = threadIdx.x + blockIdx.x * blockDim.x;

    if (idx < n){
        if (*ans < dp[ (k - 1) * n + idx]){
            atomicMax(ans, dp[ (k - 1) * n + idx]);
        }
    }
}

int MaxIncreasingSub(int *arr, int n, int k) 
{
	int *dp = (int *)malloc(n * k * sizeof(int));
    for (int i = 0; i < n * k; i++) {
        dp[i] = -1;
    }

    for (int i = 0; i < n; i++) {
        dp[i] = arr[i];
    }

    int *d_arr, *d_dp;

    cudaMalloc((void **)&d_arr, n * sizeof(int));
    cudaMalloc((void **)&d_dp, n * k * sizeof(int));
    cudaMemcpy(d_arr, arr, n * sizeof(int), cudaMemcpyHostToDevice);
    cudaMemcpy(d_dp, dp, n * k * sizeof(int), cudaMemcpyHostToDevice);
    free(dp);

	for(int i = 0; i < (k - 1); i ++){
        int threadsPerBlock = 1024;
        int totalOperations = (n * n + (i + 1) - (n + (i + 1) * (i + 1))) / 2;
        int blocksPerGrid = (totalOperations + threadsPerBlock - 1) / threadsPerBlock;
        calculatePartialSum<<<blocksPerGrid, threadsPerBlock>>>( totalOperations, i, n, d_arr, d_dp);
        cudaDeviceSynchronize();
	}
    cudaFree(d_arr);

    int ans, *d_ans;

    ans = -1;
    cudaMalloc((void **)&d_ans, sizeof(int));
    cudaMemcpy(d_ans, &ans, sizeof(int), cudaMemcpyHostToDevice);

    int threadPerBlock = 1024;
    int blockPerGrid = ( n + threadPerBlock - 1 ) / threadPerBlock;
    calculateAnswer<<<blockPerGrid, threadPerBlock>>>(n, k, d_ans, d_dp);
    cudaDeviceSynchronize();
    cudaMemcpy(&ans, d_ans, sizeof(int), cudaMemcpyDeviceToHost);
    cudaFree(d_dp);
    cudaFree(d_ans);

    return (ans == -1) ? 0 : ans; 
} 

int main() 
{ 
	int n, k;
    scanf(" %d", &n);
    scanf(" %d", &k);
    int *arr = (int *) malloc(n * sizeof(int));
	for (int i = 0; i<n; i++)
		scanf(" %d", &arr[i]);
	int ans = MaxIncreasingSub(arr, n, k); 
	printf("%d\n", ans); 
	return 0;
} 
```

## Programa paralelo com MPI

Explorando a mesma implementação dos programas anteriores em que o algoritmo sequencial foi modificado possuindo uma pior eficiência, mas nenhuma dependência entre os valores, a implementação com a biblioteca MPI, devido ao número de núcleos de processamento serem relativamente pequeno, comparando com a implementação com CUDA, as operações serão divididas em índices do vetor que deverá ser calculado a cada rodada do loop principal. Por exemplo:

O *for loop* principal, que itera dentre $1$ até $k$, a cada iteração será feito necessário calculas as somas parciais máximas de $n - i$ elementos do vetor, sendo $i$ o valor que está sendo iterado no *for loop*. Desta maneira, caso temos disponíveis $4$ núcleos de processamento, cada núcleo calculará $\dfrac{n - i}{4}$ para todos os elementos, e os primeiros $(n - i)\ \%\ 4$ núcleos receberão $1$ elemento a mais para que não sobre nenhum elemento sem ser calculado a soma parcial máxima. 

Deve se atentar que para calcular a soma parcial máxima de cada elemento do vetor é necessário fazer comparações com todos os elementos que estão posições anteriores a ele, assim os primeiros elementos serão mais fáceis de serem calculados, o que poderia criar uma inconsistência, pois os núcleos de processamento com *ranks* maiores teriam muito mais trabalho do que os de *ranks* menores. Para contornar isto, a cada elemento que é atribuído ao núcleo de processamento o próximo elemento será da posição mais alta do vetor que ainda não foi atribuída a um núcleo de processamento.

Além disso a implementação utiliza o sistema de *Master and Workers*, em que os *workers* receberão uma lista de índices do vetor, dividida e definida com as ideias apresentadas acima, e assim calculará a soma parcial máxima para o dado índice, e retornará os resultados ao *master* que ficará encarregado de administrar as somas parciais retornadas dos *workers*, salvando nos índices corretos e redistribuindo a nova configuração de somas parciais aos workers para calcularem novamente as somas máximas parciais dos índices que foram determinados a eles. Até a iteração do *loop* de $k - 1$, pois após a iteração de $k - 1$, é necessário receber o maior valor de soma parcial calculado até o momento, nesta conjuntura os *workers* irão calcular a soma parcial máxima de todos os índices que foram calculados por ele e os *workers* e *master* irão fazer a chamada de um `MPI_Reduce()` para que o master receba a maior soma parcial calculada com $k$ elementos e deste modo é mostrado ao usuário o valor retornado.

```python
%%writefile mpi-prog.c
#include <stdio.h>
#include <stdlib.h>
#include <mpi.h>
#define MAX(a, b) a > b ? a : b

typedef struct infoNode
{
	int *indexes;
	int qtdTasks;
} infoNode;

void resetArray(int tam, int *arr)
{
	for (int i = 0; i < tam; i++)
	{
		arr[i] = 0;
	}
}

void masterFunction(int totalNodes)
{
	int n, k, qtdTaskPerNode, rest, ans, maxTasks;

	scanf(" %d", &n);
	scanf(" %d", &k);

	int *arr = malloc(n * sizeof(int));
	int *dp = malloc(n * sizeof(int));
	infoNode *infosNodes = malloc((totalNodes - 1) * sizeof(infoNode));

	for (int i = 0; i < n; i++)
	{
		scanf(" %d", &arr[i]);
		dp[i] = arr[i];
	}

	// Envia o vetor a todos os nós pois será utilizado por todos para leitura.
	MPI_Bcast((void *)&n, 1, MPI_INT, 0, MPI_COMM_WORLD);
	MPI_Bcast((void *)&k, 1, MPI_INT, 0, MPI_COMM_WORLD);
	MPI_Bcast((void *)arr, n, MPI_INT, 0, MPI_COMM_WORLD);

	maxTasks = (n) / (totalNodes - 1) + 1;

	// Inicializa o vetor de indices de cada nó.
	for (int i = 0; i < totalNodes - 1; i++)
	{
		infosNodes[i].indexes = malloc(maxTasks * sizeof(int));
	}

	for (int i = 1; i < k; i++)
	{
		qtdTaskPerNode = (n - i) / (totalNodes - 1);
		rest = (n - i) % (totalNodes - 1);
		int qtdTasksSended = 0;

		MPI_Bcast(dp, n, MPI_INT, 0, MPI_COMM_WORLD);

		for (int j = 1; j < totalNodes; j++)
		{
			int qtdTask = qtdTaskPerNode;

			if (i < rest)
			{
				qtdTask += 1;
			}

			infosNodes[j - 1].qtdTasks = qtdTask;

			for (int k = 0; k < qtdTask; k++)
			{
				if (qtdTasksSended % 2 == 0)
				{
					infosNodes[j - 1].indexes[k] = i + qtdTasksSended++ / 2;
				}
				else
				{
					infosNodes[j - 1].indexes[k] = n - 1 - qtdTasksSended++ / 2;
				}
			}
			MPI_Send((void *)&qtdTask, 1, MPI_INT, j, 0, MPI_COMM_WORLD);
			MPI_Send((void *)infosNodes[j - 1].indexes, infosNodes[j - 1].qtdTasks, MPI_INT, j, 0, MPI_COMM_WORLD);
		}

		if (i < k - 1)
		{
			int *receivedNumbers = malloc(maxTasks * sizeof(int));

			resetArray(n, dp);

			for (int j = 1; j < totalNodes; j++)
			{
				MPI_Recv(receivedNumbers, infosNodes[j - 1].qtdTasks, MPI_INT, j, 0, MPI_COMM_WORLD, NULL);

				for (int k = 0; k < infosNodes[j - 1].qtdTasks; k++)
				{
					dp[infosNodes[j - 1].indexes[k]] = receivedNumbers[k];
				}
			}
			free(receivedNumbers);
		}
		else
		{
			int max = 0;
			MPI_Reduce(MPI_IN_PLACE, &ans, 1, MPI_INT, MPI_MAX, 0, MPI_COMM_WORLD);
			if (ans == -1)
			{
				ans = 0;
			}
		}
		MPI_Barrier(MPI_COMM_WORLD);
	}

	printf("%d\n", ans);
	free(arr);
	free(dp);

	for (int i = 0; i < totalNodes - 1; i++)
	{
		free(infosNodes[i].indexes);
	}

	free(infosNodes);
}

void workerFunction(int rank, int totalNodes)
{
	int *arr, n, k, *dp, *results, *indexes, maxTasks;
	int max = -1;

	// Recebe os valores do master.
	MPI_Bcast((void *)&n, 1, MPI_INT, 0, MPI_COMM_WORLD);
	MPI_Bcast((void *)&k, 1, MPI_INT, 0, MPI_COMM_WORLD);

	arr = malloc(n * sizeof(int));
	dp = malloc(n * sizeof(int));
	maxTasks = (n) / (totalNodes - 1) + 1;

	MPI_Bcast((void *)arr, n, MPI_INT, 0, MPI_COMM_WORLD);

	indexes = (int *)malloc(maxTasks * sizeof(int));
	results = (int *)calloc(maxTasks, sizeof(int));

	for (int i = 1; i < k; i++)
	{
		int qtdTasks;

		MPI_Bcast(dp, n, MPI_INT, 0, MPI_COMM_WORLD);
		MPI_Recv(&qtdTasks, 1, MPI_INT, 0, 0, MPI_COMM_WORLD, NULL);
		MPI_Recv(indexes, qtdTasks, MPI_INT, 0, 0, MPI_COMM_WORLD, NULL);

		for (int j = 0; j < qtdTasks; j++)
		{
			for (int l = 0; l < indexes[j]; l++)
			{
				if (arr[l] < arr[indexes[j]])
				{
					if (dp[l] != 0)
					{
						results[j] = MAX(results[j], dp[l] + arr[indexes[j]]);

						if (i == k - 1)
						{
							max = MAX(max, results[j]);
						}
					}
				}
			}
		}

		if (i < k - 1)
		{
			MPI_Send((void *)results, qtdTasks, MPI_INT, 0, 0, MPI_COMM_WORLD);
			resetArray(maxTasks, results);
		}
		else
		{
			MPI_Reduce(&max, NULL, 1, MPI_INT, MPI_MAX, 0, MPI_COMM_WORLD);
		}

		MPI_Barrier(MPI_COMM_WORLD);
	}

	free(results);
	free(indexes);
	free(arr);
	free(dp);
}

int main(int argc, char *argv[])
{
	int totalNodes, rank;
	int status = MPI_Init(&argc, &argv);

	if (status != MPI_SUCCESS)
	{
		printf("Erro ao iniciar o ambiente paralelo com MPI.\n");
		MPI_Abort(MPI_COMM_WORLD, status);
		return -1;
	}

	MPI_Comm_size(MPI_COMM_WORLD, &totalNodes);
	MPI_Comm_rank(MPI_COMM_WORLD, &rank);

	if (rank == 0)
	{
		masterFunction(totalNodes);
	}
	else
	{
		workerFunction(rank, totalNodes);
	}

	MPI_Finalize();

	return 0;
}
```
